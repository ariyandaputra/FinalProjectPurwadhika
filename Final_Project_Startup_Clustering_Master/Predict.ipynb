```python
Cluster_name = []

for i in range(len(DF_CLUSTER)):
    if DF_CLUSTER['Cluster'].loc[i]==0:
        Cluster_name.append('Angel Startup')
    elif DF_CLUSTER['Cluster'].loc[i]==1:
        Cluster_name.append('Big Startup')
    elif DF_CLUSTER['Cluster'].loc[i]==2:
        Cluster_name.append('Venture Startup') 
    elif DF_CLUSTER['Cluster'].loc[i]==3:
        Cluster_name.append('Early to Mid Fund Startup') 
    elif DF_CLUSTER['Cluster'].loc[i]==4:
        Cluster_name.append('Late Fund Startup') 
    elif DF_CLUSTER['Cluster'].loc[i]==5:
        Cluster_name.append('Small Startup') 
```

```python
DF_CLUSTER['Cluster_name'] = Cluster_name
```

```python
DF_CLUSTER.tail(3)
```

```python
from sklearn.model_selection import train_test_split

DFX = DF_CLUSTER[['FUND_COUNT','RAISED_USD_MEAN','COUNTRY_CODE_NO','SERIES A', 'SERIES B', 'SERIES C+', 'ANGEL', 'VENTURE', 'POST IPO', 'OTHER']]
DFY = DF_CLUSTER['Cluster']
PENGACAK = 101
PEMBAGI = 0.3

x_train,x_test,y_train,y_test = train_test_split(DFX , DFY , test_size = PEMBAGI , random_state = PENGACAK)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
```

```python
DF_USED = DF_CLUSTER
DF_NAME = 'Startup Cluster'
DFX = DF_CLUSTER[['FUND_COUNT','RAISED_USD_MEAN','COUNTRY_CODE_NO','SERIES A', 'SERIES B', 'SERIES C+', 'ANGEL', 'VENTURE', 'POST IPO', 'OTHER']]
DFY = DF_CLUSTER['Cluster']

RANDOMSTATE = 101

MODELS = [DecisionTreeClassifier(random_state=RANDOMSTATE),
          AdaBoostClassifier(base_estimator = DecisionTreeClassifier(random_state=RANDOMSTATE), algorithm='SAMME',random_state=RANDOMSTATE),
          GradientBoostingClassifier(random_state=RANDOMSTATE),
          XGBClassifier(random_state=RANDOMSTATE),
          RandomForestClassifier(random_state=RANDOMSTATE),
          AdaBoostClassifier(base_estimator = RandomForestClassifier(random_state=RANDOMSTATE), algorithm='SAMME', random_state=RANDOMSTATE),
        ]
          
    

NAMES = ['Decision Tree',
         'Adaboost - DT',
         'Gradient Boost',
         'Extreme Gradient Boost',
         'Random Forest Classifier',
         'Adaboost - RFC'
        ]

print('DF USED : ' + DF_NAME)
print('\n')
print('Y USED : ' + DFY.name)
print('\n')
print('X USED : ')
for i in DFX:
    print('- ' +i)
print('\n')
print('Model Used : ')
for i in NAMES:
    print('- ' + i)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score, recall_score, roc_auc_score

ACC = []

print('Default model validation accuracies for the train data:')
for name, model in zip(NAMES, MODELS):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test) 
    acc = accuracy_score(y_pred, y_test)
    ACC.append(acc)
    print(name + ' : ' + str(acc*100))
```
